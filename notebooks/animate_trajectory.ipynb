```python
import numpy as np
import matplotlib.pyplot as plt

from helicopter.simulation.pid import Helicopter, PIDGains
from helicopter.simulation.animator import SimulationAnimator
```

```python
throttle_gains = PIDGains(8.0, 1.5, 1.0)
pitch_gains = PIDGains(2.1, 0.001, 0.70)
yaw_gains = PIDGains(0.9, 0.05, 0.2)
helicopter = Helicopter(
    throttle_gains=throttle_gains,
    pitch_gains=pitch_gains,
    yaw_gains=yaw_gains
                        )
helicopter.loiter_radius = 5e-2

_t_span = (0., 5.)
_t_eval = np.linspace(_t_span[0], _t_span[1], int((_t_span[1] - _t_span[0]) * 25))
_setpoint = np.array([0.5, 0.5, 0.5])

_df = helicopter.solve(_t_span, _t_eval, _setpoint)
```

```python
%matplotlib ipympl

plt.close('all')
animator = SimulationAnimator()
ani = animator.plot_trajectory(_df, _setpoint)
ani
```
