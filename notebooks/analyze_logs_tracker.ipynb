```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_df(_df, column_sets, title="State Estimation", subplots: tuple[int, int] = (2, 2)):
    plt.close('all')
    fig, axes = plt.subplots(*subplots, figsize=(16, 10), sharex=True)
    fig.suptitle(title, fontsize=16)

    axes_flat = axes.flatten()

    for i, cols_key in enumerate(column_sets.keys()):
        ax = axes_flat[i]
        for _col_name in column_sets[cols_key]:
            if _col_name in _df.columns:
                ax.plot(_df['timestamp'], _df[_col_name], marker=None, label=_col_name, linewidth=1.0)
            else:
                print(f"Warning: Column '{_col_name}' not found in DataFrame.")

        ax.legend(loc='upper left')
        ax.grid(True, linestyle='--', alpha=0.6)
        ax.set_title(f"{cols_key}")

        if i >= 2:
            ax.set_xlabel("Timestamp")

    plt.tight_layout()
    plt.show()
```

```python
df = pd.read_csv('logs/tracking/log.csv')
commands_df = pd.read_csv('logs/tracking/commands_log.csv')
```

```python
col_sets = {'Position': ['px', 'py', 'pz'],
            'Quaternion': ['qw', 'qx', 'qy', 'qz'],
            'Commands': ['actual_throttle', 'actual_pitch', 'actual_yaw'],
            'Velocity': ['vx', 'vy', 'vz'],
            'Angular_Velocity': ['ox', 'oy', 'oz'],
            'Aircraft State': ['battery', 'trim']}
```

```python
plot_df(df, col_sets , subplots=(2, 3))
```

```python
command_col_sets = {'Commands': ['thrust', 'pitch', 'yaw']}
```

```python
plot_df(commands_df, command_col_sets, title="Commands", subplots=(2, 1))
```
