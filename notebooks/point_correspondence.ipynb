```python
import numpy as np
import quaternionic

import matplotlib.pyplot as plt
from helicopter.vision.test_scripts.point_correspondence import generate_reference_points, get_sample_points
from helicopter.vision.tracking import TrianglePointMatcher
```

```python
rng = np.random.default_rng(42)
ref = generate_reference_points(rng, num_reference_points=20)
```

```python
def plot_match(p_q: np.ndarray, p_t: np.ndarray, _ref: np.ndarray, _sample: np.ndarray):
    p_q = quaternionic.array(p_q)
    transformed_reference = p_q.rotate(_ref) + p_t

    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(projection='3d')


    ax.scatter(ref[:, 0], ref[:, 1], ref[:, 2], color='k')
    ax.scatter(transformed_reference[:, 0], transformed_reference[:, 1], transformed_reference[:, 2], color='b')
    ax.scatter(_sample[:, 0], _sample[:, 1], _sample[:, 2], color='r')

    ax.set_xlim(-4, 4)
    ax.set_ylim(-4, 4)
    ax.set_zlim(-4, 4)

    plt.show()
```

```python
sample_clean, _noise, sample, quat, t, corr = get_sample_points(rng, ref, noise_scale=1e-2, num_sample_points=8)
```

```python
%matplotlib ipympl

matcher = TrianglePointMatcher(ref, n=5)
```

```python
predicted_quaternion, predicted_translation = matcher.get_alignment(sample)
```

```python
plot_match(predicted_quaternion, predicted_translation, ref, sample)
```
