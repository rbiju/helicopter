```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_df(_df, column_sets, title="Sensor Data Grid", subplots: tuple[int, int] = (2, 2)):
    plt.close('all')
    fig, axes = plt.subplots(*subplots, figsize=(16, 10), sharex=True)
    fig.suptitle(title, fontsize=16)

    axes_flat = axes.flatten()

    for i, cols_key in enumerate(column_sets.keys()):
        ax = axes_flat[i]
        for _col_name in column_sets[cols_key]:
            if _col_name in _df.columns:
                ax.plot(_df['timestamp'], _df[_col_name], marker=None, label=_col_name, linewidth=1.0)
            else:
                print(f"Warning: Column '{_col_name}' not found in DataFrame.")

        ax.legend(loc='upper left')
        ax.grid(True, linestyle='--', alpha=0.6)
        ax.set_title(f"{cols_key}")

        if i >= 2:
            ax.set_xlabel("Timestamp")

    plt.tight_layout()
    plt.show()
```

```python
def set_timestamps(_df):
    _df['timestamp'] = _df['timestamp'] - _df['timestamp'].min()
    return _df
```

```python
df = pd.read_csv('logs/log.csv')
imu_df = pd.read_csv('logs/imu_log.csv')
```

```python
df = set_timestamps(df)
imu_df = set_timestamps(imu_df)
```

```python
col_sets = {'Position': ['px', 'py', 'pz'],
            'Quaternion': ['qx', 'qy', 'qz'],
            'Velocity': ['vx', 'vy', 'vz'],
            'Accel_Bias': ['b_ax', 'b_ay', 'b_az'],
            'Gyro_Bias': ['b_gx', 'b_gy', 'b_gz']}
```

```python
plot_df(df, col_sets, title="State Estimation", subplots=(2, 3))
```

```python
min_T = 8.5
max_T = 9.0
subset = df[(df['timestamp'] < max_T) & (df['timestamp'] >= min_T)]
plot_df(subset, col_sets, title=f"State Estimation t=[{min_T}, {max_T}] ", subplots=(2, 3))
```
