```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('log.csv')
imu_df = pd.read_csv('imu_log.csv')
vision_df = pd.read_csv('vision_log.csv')
```

```python
df['timestamp'] = df['timestamp'] - df['timestamp'].min()
```
