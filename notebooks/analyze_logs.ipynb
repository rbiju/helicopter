```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('log.csv')
imu_df = pd.read_csv('imu_log.csv')
vision_df = pd.read_csv('vision_log.csv')
```

```python
df['timestamp'] = df['timestamp'] - df['timestamp'].min()
```

```python
df.head()
```

```python
def get_quantity_range(_df, col_name):
    return _df[col_name].max() - _df[col_name].min()
```

```python
get_quantity_range(df, 'pz')
```

```python
df['pz'].min()
```

```python
imu_subset = df[df['event'] == 'imu']
```

```python
%matplotlib ipympl
plt.close('all')

fig, ax = plt.subplots(1, 1)
for col_name in ['px', 'py', 'pz']:
    ax.plot(imu_subset['timestamp'], imu_subset[col_name], marker=None, label=col_name)

ax.legend()
plt.show()
```
