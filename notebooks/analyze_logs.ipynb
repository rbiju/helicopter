```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def plot_state(_df, column_sets, title="Sensor Data Grid"):
    if len(column_sets) != 4:
        print("Error: Please provide exactly 4 sets of columns.")
        return

    fig, axes = plt.subplots(2, 2, figsize=(14, 10), sharex=True)
    fig.suptitle(title, fontsize=16)

    axes_flat = axes.flatten()

    for i, ax in enumerate(axes_flat):
        cols_key = list(column_sets.keys())[i]
        for _col_name in column_sets[cols_key]:
            if _col_name in _df.columns:
                ax.plot(_df['timestamp'], _df[_col_name], marker=None, label=_col_name, linewidth=1.5)
            else:
                print(f"Warning: Column '{_col_name}' not found in DataFrame.")

        ax.legend(loc='upper right')
        ax.grid(True, linestyle='--', alpha=0.6)
        ax.set_title(f"{cols_key}")

        if i >= 2:
            ax.set_xlabel("Timestamp")

    plt.tight_layout()
    plt.show()
```

```python
def set_timestamps(_df):
    _df['timestamp'] = _df['timestamp'] - _df['timestamp'].min()
    return _df
```

```python
warmup_df = pd.read_csv('warmup.csv')
df = pd.read_csv('log.csv')
imu_df = pd.read_csv('imu_log.csv')
vision_df = pd.read_csv('vision_log.csv')
```

```python
warmup_df = set_timestamps(warmup_df)
df = set_timestamps(df)
imu_df = set_timestamps(imu_df)
vision_df = set_timestamps(vision_df)
```

```python
col_sets = {'Position': ['px', 'py', 'pz'],
            'Accel_Bias': ['b_ax', 'b_ay', 'b_az'],
            'Gyro_Bias': ['b_gx', 'b_gy', 'b_gz'],
            'Quaternion': ['qx', 'qy', 'qz']}
```

```python
plot_state(warmup_df, col_sets, title="Warmup")
```

```python
plot_state(df, col_sets, title="State Estimation")
```
