```python
import numpy as np
from scipy.spatial.transform import Rotation

import matplotlib.pyplot as plt
```

```python
def plot_3d(rotation: Rotation, window=2, marker_size=10, c=None, show_indices=True):
    plt.close('all')

    fig = plt.figure(figsize=(8, 8))
    ax = fig.add_subplot(projection='3d')

    points = np.array([
        [1, 1, 0],
        [ 1, -1, 0],
        [-1, -1, 0],
        [-1,  1, 0]
    ])

    x = points[:, 0]
    y = points[:, 1]
    z = points[:, 2]

    if c is None:
        c = ['red', 'green', 'blue', 'yellow']

    ax.scatter(x, y, z, c=c, s=marker_size, depthshade=True)

    if show_indices:
        for i in range(len(points)):
            ax.text(x[i], y[i], z[i], str(i), size=12, color='black', zorder=10)

    if rotation is not None:
        rotated_points = rotation.apply(points)
        x_rot = rotated_points[:, 0]
        y_rot = rotated_points[:, 1]
        z_rot = rotated_points[:, 2]

        ax.scatter(x_rot, y_rot, z_rot, c=c, s=marker_size, depthshade=True, alpha=0.7)

    ax.set_xlim(-window, window)
    ax.set_ylim(-window, window)
    ax.set_zlim(-window, window)

    ax.set_xlabel('X')
    ax.set_ylabel('Y')
    ax.set_zlabel('Z')

    plt.show()
```

```python
marker_dict = {1: {'position': np.array([ 1.14592211,  0.37522784, -0.47579283]),
                   'orientation': np.array([ -3.52399818,  -8.94130208, -16.5152128 ]),
                   'offset': Rotation.from_euler('z', [90], degrees=True)},
               0: {'position': np.array([ 1.31232266,  0.41695576, -0.50475312]),
                   'orientation': np.array([     81.437,      52.032,     -63.063]),
                   'offset': Rotation.from_euler('y', [90], degrees=True)}}
```

```python
position_offset = np.array([-0.355,-0.685, 0.0])
```

```python
camera_quat = Rotation.from_rotvec(np.array([-0.31031, 15.377, 0]), degrees=True)
```

```python
def compute_yaw_rotation(key: int):
    marker_quat = Rotation.from_rotvec(marker_dict[key]['orientation'], degrees=True)
    world_space_marker_rotation = camera_quat * marker_quat * marker_dict[key]['offset'].inv()
    yaw_only_marker_quat = Rotation.from_euler('z',
                                           world_space_marker_rotation.as_euler('zyx')[0])

    return yaw_only_marker_quat
```

```python
yaw_only_0 = compute_yaw_rotation(0)
yaw_only_1 = compute_yaw_rotation(1)
```

```python
yaw_only_0.as_rotvec(degrees=True)
```

```python
yaw_only_1.as_rotvec(degrees=True)
```

```python
origin_quat_0 = camera_quat * yaw_only_0
origin_quat_1 = camera_quat * yaw_only_1
```

```python
origin_pos_0 = camera_quat.apply(marker_dict[0]['position']) - yaw_only_0.apply(position_offset)
origin_pos_1 = camera_quat.apply(marker_dict[1]['position']) - yaw_only_1.apply(position_offset)
```

```python
origin_quat_0.apply(np.array([0.5, 0, 0.])) + origin_pos_0
```

```python
origin_quat_0.inv().apply(np.array([ 1.72165225, -0.51095129, -0.81984313]) - origin_pos_0)
```
